```python
def price_to_book_value(price, book_value):
    if len(book_value) == 1:
        return price / book_value[0]
    else:
        bookValue = book_value[0] / book_value[1]
        return price / bookValue

price = int(input("Masukkan harga saham: "))
book_value = list(map(int, input("Masukkan harga BVPS atau banyak saham dan nilai ekuitas: ").split()))
print(price_to_book_value(price, book_value))
```

```python
def PBV_BVPS(market_price, bvps):
  return market_price / bvps
def PBV_Equity(market_price, equity, stocks):
  return market_price / (equity / stocks)
```

```python
def quick_sort(arr):
  if len(arr) <= 1:
      return arr

  pivot = arr[-1]

  left = [x for x in arr[:-1] if x <= pivot]
  right = [x for x in arr[:-1] if x > pivot]

  return quick_sort(right) + [pivot] + quick_sort(left)
```

```python
data = pd.read_csv('stocks_data.csv')
data.head()
```
